```python
from tmtoolkit.lda_utils.common import get_doc_lengths, get_most_salient_words, get_least_salient_words, \
    generate_topic_labels_from_top_words

doc_lengths = get_doc_lengths(dtm)
get_most_salient_words(vocab, model.topic_word_, model.doc_topic_, doc_lengths, n=30)
```

```python
get_least_salient_words(vocab, model.topic_word_, model.doc_topic_, doc_lengths, n=30)
```

```python
from tmtoolkit.lda_utils.common import get_most_distinct_words, get_least_distinct_words

get_most_distinct_words(vocab, model.topic_word_, model.doc_topic_, doc_lengths, n=30)
```

```python
get_least_distinct_words(vocab, model.topic_word_, model.doc_topic_, doc_lengths, n=50)
```

```python
named_topic_labels = generate_topic_labels_from_top_words(model.topic_word_, model.doc_topic_, doc_lengths, vocab, lambda_=0.6, n_words=2)
named_topic_labels
```

```python
import numpy as np

#topic_indices = np.array([68, 20, 52, 21, 41])-1
topic_indices = np.array([16, 47, 35, 36, 28])-1
doc_indices = list(reversed(np.argsort(model.doc_topic_[:, topic_indices].sum(axis=1))))[:7]
```

```python
import matplotlib.pyplot as plt
from tmtoolkit.lda_utils.visualize import plot_doc_topic_heatmap

fig, ax = plt.subplots(figsize=(8, 8))  # create a figure to draw on

plot_doc_topic_heatmap(fig, ax, model.doc_topic_, doc_labels,
                       topic_labels=named_topic_labels,
                       which_document_indices=doc_indices,
                       which_topic_indices=topic_indices,
                       round_values_in_cells=3,   # round to 3 digits after point
                       fontsize_cell_values=8)    # set smaller font size in cells

plt.tight_layout()  # make sure the tick and axes labels are correctly displayed
fig
```
