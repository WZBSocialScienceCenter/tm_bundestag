```python
from tmtoolkit.topicmod.model_stats import generate_topic_labels_from_top_words

named_topic_labels = generate_topic_labels_from_top_words(model.topic_word_, model.doc_topic_, doc_lengths, vocab, lambda_=0.6, n_words=2)
named_topic_labels
```

```python
import numpy as np

topic_indices = np.array([16, 21, 25])-1
doc_indices = list(reversed(np.argsort(model.doc_topic_[:, topic_indices].sum(axis=1))))[:7]
```

```python
import matplotlib.pyplot as plt
from tmtoolkit.topicmod.visualize import plot_doc_topic_heatmap

fig, ax = plt.subplots(figsize=(8, 8))  # create a figure to draw on

plot_doc_topic_heatmap(fig, ax, model.doc_topic_, doc_labels,
                       topic_labels=named_topic_labels,
                       which_document_indices=doc_indices,
                       which_topic_indices=topic_indices,
                       round_values_in_cells=3,   # round to 3 digits after point
                       fontsize_cell_values=8)    # set smaller font size in cells

plt.tight_layout()  # make sure the tick and axes labels are correctly displayed
plt.show();
```

## Topic coherence

### $U_{mass}$ metric (Mimno et al. 2011)

```python
from tmtoolkit.topicmod.evaluate import metric_coherence_mimno_2011

coh = metric_coherence_mimno_2011(model.topic_word_, dtm, top_n=20)   # use top 20 words per topic for metric
coh
```

```python
import matplotlib.pyplot as plt

plt.hist(coh, bins=20)
plt.xlabel('coherence')
plt.ylabel('n')
plt.show();
```

```python
np.median(coh), np.mean(coh), np.std(coh)
```

```python
import numpy as np

top10_t_indices = np.argsort(coh)[::-1][:10]
bottom10_t_indices = np.argsort(coh)[:10]

print('top 10 topics (with best coherence values):')
print('-------------------------------------------')
for t_ind, t_coh in zip(top10_t_indices, coh[top10_t_indices]):
    print('%s (coherence=%f)' % (named_topic_labels[t_ind], t_coh))

print('')
    
print('bottom 10 topics (with worst coherence values):')
print('----------------------------------------------')
for t_ind, t_coh in zip(bottom10_t_indices, coh[bottom10_t_indices]):
    print('%s (coherence=%f)' % (named_topic_labels[t_ind], t_coh))
```

### $C_V$ metric (Röder et al. 2015)

```python
from tmtoolkit.topicmod.evaluate import metric_coherence_gensim

_, _, _, tokens = unpickle_file('data/speeches_tokens_2.pickle')

coh_c_v = metric_coherence_gensim('c_v', model.topic_word_, dtm=dtm, vocab=vocab,
                                  texts=[list(tokens[dl]) for dl in doc_labels],
                                  top_n=20)   # use top 20 words per topic for metric
coh_c_v
```

```python
import matplotlib.pyplot as plt

plt.hist(coh_c_v, bins=40)
plt.xlabel('coherence')
plt.ylabel('n')
plt.show();
```

```python
import numpy as np

coh_c_v = np.array(coh_c_v)

top10_t_indices = np.argsort(coh_c_v)[::-1][:10]
bottom10_t_indices = np.argsort(coh_c_v)[:10]

print('top 10 topics (with best coherence values):')
print('-------------------------------------------')
for t_ind, t_coh in zip(top10_t_indices, coh_c_v[top10_t_indices]):
    print('%s (coherence=%f)' % (named_topic_labels[t_ind], t_coh))

print('')
    
print('bottom 10 topics (with worst coherence values):')
print('----------------------------------------------')
for t_ind, t_coh in zip(bottom10_t_indices, coh_c_v[bottom10_t_indices]):
    print('%s (coherence=%f)' % (named_topic_labels[t_ind], t_coh))
```
